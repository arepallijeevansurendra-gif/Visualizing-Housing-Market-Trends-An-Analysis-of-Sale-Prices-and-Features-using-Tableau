# Housing Market Analysis

This notebook prepares the housing dataset, performs cleaning, explores features, and exports a ready dataset for Tableau.

```python
import pandas as pd
from pathlib import Path

# Update this path to match your dataset location.
data_path = Path(r"C:\Users\Vineela\Desktop\Data analyst\data\housing_data.csv")
df = pd.read_csv(data_path)

# Review basic dataset structure
print('rows, cols:', df.shape)
print(df.dtypes)
print(df.head())
```

```python
# Standardize column names and inspect missing values
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('.', '', regex=False)
missing = df.isna().sum()
print(missing[missing > 0])

# Rename fields if needed for readability
rename_map = {
    'yr_renovated': 'year_renovated',
    'yr_built': 'year_built',
    'sqft_living15': 'sqft_living_15',
    'sqft_lot15': 'sqft_lot_15'
}
df = df.rename(columns=rename_map)
print(df.columns.tolist())
```

```python
# Create calculated fields
current_year = 2026
df['house_age'] = current_year - df['year_built']
df['years_since_renovation'] = df['year_renovated'].replace(0, pd.NA)
df['years_since_renovation'] = df['years_since_renovation'].apply(lambda x: current_year - x if pd.notna(x) else x)
df['renovated'] = df['year_renovated'] > 0
df['renovation_status'] = df['renovated'].map({True: 'Renovated', False: 'Not Renovated'})

# Validate key distributions
summary = df[['price', 'house_age', 'years_since_renovation', 'bathrooms', 'bedrooms', 'floors']].describe()
print(summary)
```

```python
# Optional cleaning: drop rows without price or primary features
required_columns = ['price', 'house_age', 'bathrooms', 'bedrooms', 'floors']
df = df.dropna(subset=required_columns)

# Export cleaned dataset for Tableau use
output_path = Path('..') / 'data' / 'housing_data_cleaned.csv'
df.to_csv(output_path, index=False)
print('Exported cleaned dataset to', output_path)
```
